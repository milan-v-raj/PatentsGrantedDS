 **HANDLING INCONSISTENT APPLICATION NUMBER FORMATTING FOR UNIFORMITY AND IMPROVED DATA RETRIEVAL**

 (**CONVERTING** **TO** **FORMAT** **NUM**/**IPO** **LOCATION**/**YEAR** **OF** **FILING**)

```python
# Function to check if the application number is in the old format
def is_old_format(app_number):
    parts = app_number.split('/')
    if len(parts) == 3:  # Should have three parts
        unique_id, location_code, year = parts
        # Check if the year is four digits and the location code is valid
        return (len(year) == 4 and location_code in ['DEL', 'MUM', 'KOL', 'CHE', 'CHENP', 'DELNP', 'MUMNP', 'KOLNP'])
    return False

# Function to convert new application number to old format
def convert_to_old_format(new_number):
    # Extract year, location digit, and unique ID
    year = new_number[:4]        # First four digits for the year
    location_digit = new_number[4]  # Fifth digit for the location code
    unique_id = new_number[8:]   # Last digits for the unique ID

    # Map the location digit back to the corresponding code
    location_code = {
        '1': 'DEL',
        '2': 'MUM',
        '3': 'KOL',
        '4': 'CHE'
    }.get(location_digit, 'UNKNOWN')  # Handle invalid location as needed

    # Create the old application number format
    old_application_number = f"{unique_id}/{location_code}/{year}"

    return old_application_number

# Main function to process the DataFrame and standardize application numbers
def process_application_numbers(df):
    # Apply the conversion function only if the number is not already in old format
    df['APPLICATION_NUMBER'] = df['APPLICATION_NUMBER'].apply(
        lambda x: convert_to_old_format(x) if not is_old_format(x) else x
    )
    return df

# Example usage:
# modified_df = process_application_numbers(input_df)

modified_df=process_application_numbers(df)
display(modified_df)
```

```python
# Convert date columns to datetime format with dayfirst=True
df['PUBLICATION_DATE'] = pd.to_datetime(df['PUBLICATION_DATE'], dayfirst=True, errors='coerce')
df['DATE_OF_FILING'] = pd.to_datetime(df['DATE_OF_FILING'], dayfirst=True, errors='coerce')

# Calculate the lag (in days) between filing and publication
df['Filing_to_Publication_Lag'] = (df['PUBLICATION_DATE'] - df['DATE_OF_FILING']).dt.days
```

**Outlier Treatment**

```python
# Function to find the number of outliers in a column with reduced sensitivity
def count_outliers(column, multiplier=2.5):  # Adjust multiplier for reduced sensitivity
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers), lower_bound, upper_bound

# Apply the function to each numerical column with a higher threshold
numerical_columns = ['NO_OF_PAGES', 'NO_OF_CLAIMS', 'Filing_to_Publication_Lag']  # Adjust with numerical columns in your dataset
outlier_counts = {col: count_outliers(col, multiplier=1.5) for col in numerical_columns}

# Display the results
print("Number of outliers in each column with reduced sensitivity (multiplier=2.5):")
for col, (count, lower, upper) in outlier_counts.items():
    print(f"{col}: {count} outliers (Thresholds: {lower} to {upper})")

```

**Checking if the filing to publication lag days are normally distributed for different fields  (shapiro and anderson darling(as dats is big))**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, kstest, norm, probplot, anderson

# Set a sample size for the Shapiro-Wilk test
sample_size = 500

# Perform normality checks and plot Q-Q plots for each FIELD_OF_INVENTION
fields = df['FIELD_OF_INVENTION'].unique()

# Set up subplots for Q-Q plots
num_fields = len(fields)
fig, axes = plt.subplots(nrows=(num_fields + 2) // 3, ncols=3, figsize=(15, 4 * ((num_fields + 2) // 3)))
axes = axes.flatten()  # Flatten the axes array for easy iteration

normality_results = {}  # Dictionary to store normality test results

for i, field in enumerate(fields):
    # Extract data for the current field
    field_data = df[df['FIELD_OF_INVENTION'] == field]['Filing_to_Publication_Lag'].dropna()

    # Sample data if it's larger than the sample_size
    if len(field_data) > sample_size:
        field_data = field_data.sample(sample_size, random_state=0)

    # Perform Shapiro-Wilk test for normality
    shapiro_stat, shapiro_p = shapiro(field_data)

    # Perform Anderson-Darling test (better for larger samples)
    anderson_result = anderson(field_data, dist='norm')

    # Store results
    normality_results[field] = {
        'Shapiro-Wilk p-value': shapiro_p,
        'Anderson-Darling statistic': anderson_result.statistic,
        'Anderson-Darling critical values': anderson_result.critical_values
    }

    # Generate Q-Q plot
    probplot(field_data, dist="norm", plot=axes[i])
    axes[i].set_title(f'Q-Q Plot for {field} (SW p={shapiro_p:.3f})')

# Adjust layout
plt.tight_layout()
plt.suptitle('Q-Q Plots by Field of Invention', y=1.02)
plt.show()

# Print the normality test results
print("Normality Test Results:")
for field, results in normality_results.items():
    print(f"{field}:")
    for test, value in results.items():
        print(f"  {test}: {value}")
    print()
```

**Mann-Whitney U Test: For comparing the number of claims between two IPO locations, especially if the data is non-normal.**

```python
from itertools import combinations
from scipy.stats import mannwhitneyu

# Get all unique IPO locations
locations = df['IPO_LOCATION'].dropna().unique()

# Initialize a dictionary to store the results
mannwhitney_results = {}

# Loop through all unique pairs of locations
for loc1, loc2 in combinations(locations, 2):
    # Extract data for each location, removing NaN values
    loc1_data = df[df['IPO_LOCATION'] == loc1]['NO_OF_CLAIMS'].dropna()
    loc2_data = df[df['IPO_LOCATION'] == loc2]['NO_OF_CLAIMS'].dropna()

    # Check if both groups have data
    if len(loc1_data) > 0 and len(loc2_data) > 0:
        # Perform Mann-Whitney U test
        u_statistic, p_value = mannwhitneyu(loc1_data, loc2_data, alternative='two-sided')
        # Store the result in the dictionary
        mannwhitney_results[(loc1, loc2)] = {'U statistic': u_statistic, 'p-value': p_value}
    else:
        mannwhitney_results[(loc1, loc2)] = "Insufficient data"

# Display the results
for pair, result in mannwhitney_results.items():
    if result != "Insufficient data":
        print(f"{pair[0]} vs {pair[1]}: U statistic = {result['U statistic']}, p-value = {result['p-value']}")
    else:
        print(f"{pair[0]} vs {pair[1]}: {result}")
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['FIELD_OF_INVENTION'], df['Filing_to_Publication_Lag'])

# Perform chi-squared test of independence
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-squared = {chi2}, p-value = {p_value}, degrees of freedom = {dof}")
```

**Chi-Squared Test:**

**Statistical Significance:** The p-value is 0.0, which is less than the conventional significance level of 0.05. This indicates that the association between the FIELD_OF_INVENTION and Filing_to_Publication_Lag is statistically significant. In other words, there is strong evidence that the two variables are not independent.
**Strength of Association: **The chi-squared statistic is very large, further confirming a strong association between the two variables. However, the chi-squared statistic alone doesn't provide a measure of the strength of association.

```python
import numpy as np

# Cramér's V formula
n = contingency_table.values.sum()  # Total sample size
cramer_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))

print(f"Cramér's V: {cramer_v}")
```

**Strength of Association:** Cramer's V is a measure of the strength of association between two categorical variables. It ranges from 0 to 1, with higher values indicating a stronger association.

 **t-tests for consecutive pairs of years with respect to number of claims**

```python
import pandas as pd
from scipy import stats

# Ensure 'FILING_YEAR' and 'NO_OF_CLAIMS' columns are in the dataframe
df = df.dropna(subset=['FILING_YEAR', 'NO_OF_CLAIMS'])

# Get unique years and sort them
years = sorted(df['FILING_YEAR'].unique())

results = []

# Perform t-tests for consecutive pairs of years
for i in range(len(years) - 1):
    year1 = years[i]
    year2 = years[i + 1]

    claims_year1 = df[df['FILING_YEAR'] == year1]['NO_OF_CLAIMS']
    claims_year2 = df[df['FILING_YEAR'] == year2]['NO_OF_CLAIMS']

    t_stat, p_value = stats.ttest_ind(claims_year1, claims_year2, nan_policy='omit')

    result = {
        'Year 1': year1,
        'Year 2': year2,
        'T-statistic': t_stat,
        'P-value': p_value
    }
    results.append(result)

    print(f"Comparing {year1} vs {year2}:")
    print(f"T-statistic: {t_stat}, P-value: {p_value}")

    if p_value < 0.05:
        print("Reject the null hypothesis: There is a significant difference in the number of claims between these years.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in the number of claims between these years.")
    print()

# Convert results to a DataFrame for better visualization if needed
results_df = pd.DataFrame(results)
print(results_df)
```
